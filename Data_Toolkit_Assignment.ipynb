# Data Toolkit

- 1) What is NumPy, and why is it widely used in Python?
- Answer
 - NumPy is a Python library used for numerical computing. It's widely used because it offers fast, memory-efficient operations on large arrays and matrices, and supports mathematical functions, random number generation, linear algebra, and more.

- 2) How does broadcasting work in NumPy?
- Answer
 - Broadcasting allows NumPy to perform operations on arrays of different shapes by automatically expanding them to a compatible shape without copying data.

- 3) What is a Pandas DataFrame?
- Answer
 - A DataFrame is a 2D, table-like data structure in Pandas with labeled axes (rows and columns), perfect for handling and analyzing structured data.

- 4) Explain the use of the groupby() method in Pandas.
- Answer
 - The groupby() method is used to group data based on one or more columns, then apply aggregation (like sum, mean) or transformations to each group.

- 5) Why is Seaborn preferred for statistical visualizations?
- Answer
 - Seaborn provides attractive, high-level interfaces for drawing informative statistical graphics with less code, especially for complex plots like violin, box, and distribution plots.

- 6) What are the differences between NumPy arrays and Python lists?
- Answer
 - NumPy arrays are faster and use less memory.
 - They support element-wise operations and broadcasting.
 - Python lists are more flexible but slower for numerical tasks.

- 7) What is a heatmap, and when should it be used?
- Answer
 - A heatmap is a 2D graphical representation of data where values are shown as colors. It's great for visualizing correlations or matrices, like confusion matrices or feature correlations.

- 8) What does the term “vectorized operation” mean in NumPy?
- Answer
 - Vectorized operations refer to performing operations on entire arrays without explicit loops, which makes the code faster and more concise.

- 9) How does Matplotlib differ from Plotly?
- Answer
 - Matplotlib is static and widely used in scientific computing.
 - Plotly creates interactive, web-based plots ideal for dashboards or reports.

- 10) What is the significance of hierarchical indexing in Pandas?
- Answer
 - Hierarchical indexing allows for multi-level row or column indexes, which enables working with higher-dimensional data in a 2D structure and performing more complex data manipulations.

- 11) What is the role of Seaborn’s pairplot() function?
- Answer
 - pairplot() creates a matrix of scatter plots for visualizing relationships between multiple numerical variables in a dataset.

- 12) What is the purpose of the describe() function in Pandas?
- Answer
 - The describe() function provides summary statistics (like mean, std, min, max) for each numeric column in a DataFrame.

- 13) Why is handling missing data important in Pandas?
- Answer
 - Missing data can distort analysis results. Pandas provides tools like fillna() and dropna() to handle it properly, ensuring accurate insights.

- 14) What are the benefits of using Plotly for data visualization?
- Answer
 - Plotly allows for interactive plots, supports many chart types, works in web apps, and has great integration with Dash for building dashboards.

- 15) How does NumPy handle multidimensional arrays?
- Answer
 - NumPy supports n-dimensional arrays (ndarrays), enabling efficient computation across multiple dimensions with methods like reshape(), transpose(), and indexing.

- 16) What is the role of Bokeh in data visualization?
- Answer
 - Bokeh creates interactive plots for web applications. It’s great for streaming, zooming, and linking data in real-time dashboards.

- 17) Explain the difference between apply() and map() in Pandas.
- Answer
 - map() is used for element-wise operations on Series.
 - apply() can be used on Series or DataFrames and is more flexible for row/column-wise operations.

- 18) What are some advanced features of NumPy?
- Answer
 - Memory mapping large files
 - Broadcasting rules
 - Vectorization
 - Structured arrays
 - Linear algebra and FFT support

- 19) How does Pandas simplify time series analysis?
- Answer
 - Pandas provides built-in support for datetime indexing, resampling, time shifting, and rolling windows, making time series data easy to manipulate and analyze.

- 20) What is the role of a pivot table in Pandas?
- Answer
 - A pivot table summarizes data by grouping it and applying aggregation functions. It's useful for comparing values across categories.

- 21) Why is NumPy’s array slicing faster than Python’s list slicing?
- Answer
 - NumPy arrays use contiguous memory blocks and optimized C-backend operations, making slicing and data access much faster than Python lists.

- 22) What are some common use cases for Seaborn?
- Answer
 - Visualizing distributions (histograms, KDE)
 - Analyzing relationships (scatter, pairplot)
 - Comparing categories (boxplot, barplot)
 - Heatmaps for correlations







# Practical

```python
# 1) How do you create a 2D NumPy array and calculate the sum of each row?

import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])
row_sums = np.sum(arr, axis=1)
print(row_sums)
```

```python
# 2) Write a Pandas script to find the mean of a specific column in a DataFrame

import pandas as pd

df = pd.DataFrame({'Age': [20, 25, 30, 22]})
mean_age = df['Age'].mean()
print(mean_age)
```

```python
# 3) Create a scatter plot using Matplotlib
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# 4) How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import seaborn as sns
import pandas as pd

df = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [4, 3, 2, 1],
    'C': [1, 3, 5, 7]
})
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
```

```python
# 5) Generate a bar plot using Plotly
import plotly.express as px

data = {'Fruits': ['Apple', 'Banana', 'Cherry'], 'Count': [10, 15, 7]}
fig = px.bar(data, x='Fruits', y='Count', title='Fruit Count')
fig.show()
```

```python
# 6) Create a DataFrame and add a new column based on an existing column
import pandas as pd

df = pd.DataFrame({'Scores': [50, 70, 90]})
df['Bonus'] = df['Scores'] + 10
print(df)
```

```python
# 7) Write a program to perform element-wise multiplication of two NumPy arrays
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
result = a * b
print(result)
```

```python
# 8) Create a line plot with multiple lines using Matplotlib
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [2, 3, 5, 7]
plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.legend()
plt.title("Multiple Lines")
plt.show()
```

```python
# 9)  Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold
import pandas as pd

df = pd.DataFrame({'Value': [5, 10, 15, 20]})
filtered_df = df[df['Value'] > 10]
print(filtered_df)
```

```python
# 10) Create a histogram using Seaborn to visualize a distribution
import seaborn as sns

data = [1, 2, 2, 3, 3, 3, 4, 4, 5]
sns.histplot(data, bins=5, kde=True)
```

```python
# 11) Perform matrix multiplication using NumPy
import numpy as np

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
result = np.dot(a, b)
print(result)
```

```python
# 12) Use Pandas to load a CSV file and display its first 5 rows
import pandas as pd

df = pd.read_csv('yourfile.csv')
print(df.head())
```

```python
# 13)  Create a 3D scatter plot using Plotly
import plotly.express as px
import pandas as pd

df = pd.DataFrame({
    'x': [1, 2, 3, 4],
    'y': [10, 15, 13, 17],
    'z': [5, 3, 6, 8]
})
fig = px.scatter_3d(df, x='x', y='y', z='z', title='3D Scatter Plot')
fig.show()
```
